```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
_x = np.linspace(-5, 5, 200)
```

```python
_y_relu = tf.nn.relu(_x)
_y_sigmoid = tf.nn.sigmoid(_x)
_y_tanh = tf.nn.tanh(_x)
_y_softplus = tf.nn.softplus(_x)
```

```python
sess = tf.Session()
_y_relu, _y_sigmoid, _y_tanh, _y_softplus = sess.run([_y_relu, _y_sigmoid, _y_tanh, _y_softplus])
```

```python
plt.figure(1, figsize=(8, 6))
plt.subplot(221)
plt.plot(_x, _y_relu, c='red', label='relu')
plt.ylim((-1, 5))
plt.legend(loc='best')
```

```python
plt.subplot(222)
plt.plot(_x, _y_sigmoid, c='red', label='sigmoid')
plt.ylim((-0.2, 1.2))
plt.legend(loc='best')
```

```python
plt.subplot(223)
plt.plot(_x, _y_tanh, c='red', label='tanh')
plt.ylim((-1.2, 1.2))
plt.legend(loc='best')
```

```python
plt.subplot(224)
plt.plot(_x, _y_softplus, c='red', label='softplus')
plt.ylim((-0.2, 6))
plt.legend(loc='best')

plt.show()
```
